```python
from fs.osfs import OSFS
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import requests
import urllib.request
```

```python
keywords = ['Corgi', 'Malamute', 'German shepherd', 'Golden retriever']
```

```python
for keyword in keywords:
    with OSFS('./FOLDERTESTING') as myfs:
        if (not myfs.exists(keyword)):
            myfs.makedir(keyword)
```

```python
driver = webdriver.Chrome(executable_path = 'D:\chromedriver')
```

```python
for keyword in keywords:
    driver.get('https://google.com/search?q='+keyword+'&tbm=isch'+'&tbs=isz:lt,islt:svga')
    for i in range(0,15):
        driver.execute_script("window.scrollBy(0,document.body.scrollHeight)")
        try:
            driver.find_element_by_xpath('//*[@id="islmp"]/div/div/div/div/div[2]/div[2]/input').click()
        except Exception:
            pass
        time.sleep(2)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    img_tag = soup.find_all('img', class_='rg_i')
    count = 1
    for i in img_tag:
        try:
            urllib.request.urlretrieve(i['src'], './FOLDERTESTING/'+keyword+'/'+keyword+str(count)+".jpg")
            count += 1
        except Exception as e:
            pass
```
