```python
import pandas as pd
```

```python
TLCP = pd.read_json('timeline-cases-by-provinces.json')
R12P = pd.read_json('round-1to2-by-provinces.json')
```

```python
R12P_NEW = R12P[R12P['province'] == "กรุงเทพมหานคร"]
TLCP_NEW = TLCP[TLCP['province']=="กรุงเทพมหานคร"]
BKK_CASE = R12P_NEW.append(TLCP_NEW) #Bangkok daily case dataframe
Provinces_CASE = R12P.append(TLCP) #Provinces daily case dataframe
```

```python
Thailand_CASE = Provinces_CASE.groupby(['txn_date']).sum() # grouping by date in order to produce Thailand case
```

```python
Thailand_CASE.plot(y='new_case',figsize=(8,5)) #Thailand daily graph
```

```python
BKK_CASE.plot(x='txn_date', y='new_case',figsize=(8,5))
```
